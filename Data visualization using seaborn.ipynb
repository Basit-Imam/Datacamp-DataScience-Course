# Visualizing a Categorical and a Quantitative Variable

## Count plots and barplots (Categorical plots)

```python
survey_data = pd.read_csv("young-people-survey-responses.csv")
survey_data.head()
```

```python
# Create count plot of internet usage
sns.catplot(x="Internet usage", data=survey_data, kind="count")
plt.xticks(rotation=90)


# Show plot
plt.show()
```

```python
# Create column subplots based on age category
sns.set_context("talk")
sns.catplot(x="Internet usage", data=survey_data,
            kind="count",col="Age",col_wrap=4)

# Show plot
plt.show()
```

```python
# Create a bar plot of interest in math, separated by gender
sns.catplot(x="Gender", y="Mathematics", data=survey_data, kind="bar")


# Show plot
plt.show()
```

## Box Plots

```python
#Shows the distribution of quantitative data
#median, spread, skewness & outliers
#Facilitates comparisions between groups
```

```python
#changing the whiskers using 'whis'
#by default, the whiskers extends to 1.5* interquartile range (25th to 75th % of distribution of data)
#whis=2.0 make them extend to 2.0*IQR
# whis[5,95] Lower whisker at 5th percentile & upper whisker at 95th percentile
#whis[0,100] Show min and max values
#syn = "" removes the outliers
```

```python
# Specify the category ordering
study_time_order = ["<2 hours", "2 to 5 hours", 
                    "5 to 10 hours", ">10 hours"]

# Create a box plot and set the order of the categories
sns.catplot(x="study_time", y="G3", data=student_data, kind="box",order=study_time_order)

# Show plot
plt.show()
```

```python
# Create a box plot with subgroups and omit the outliers
sns.catplot(x="internet", y="G3", data=student_data, kind="box",hue="location", sym="")


# Show plot
plt.show()
```

### Adjusting the whiskers

```python
# Set the whiskers to 0.5 * IQR
sns.catplot(x="romantic", y="G3",
            data=student_data,
            kind="box",whis=0.5)

# Show plot
plt.show()
```

```python
# Set the whiskers at the min and max values
sns.catplot(x="romantic", y="G3",
            data=student_data,
            kind="box",
            whis=[0, 100])

# Show plot
plt.show()
```

## Point Plots

```python
#Different from line plot since in line plot, x axis is quantitative variable 
#while in case of point plot it is a categorical variable
```

```python
# Create a point plot of family relationship vs. absences
sns.catplot(x="famrel", y="absences", data=student_data, kind="point")

          
# Show plot
plt.show()
```

```python
# Remove the lines joining the points
sns.catplot(x="famrel", y="absences",
            data=student_data,
            kind="point",
            capsize=0.2,join=False)
            
# Show plot
plt.show()
```

```python
# Create a point plot with subgroups
sns.catplot(x="romantic", y="absences", data=student_data, kind="point",hue="school")

# Show plot
plt.show()
```

```python
# Import median function from numpy
from numpy import median

# Plot the median number of absences instead of the mean
sns.catplot(x="romantic", y="absences",
            data=student_data,
            kind="point",
            hue="school",
            ci=None,estimator=median
            )

# Show plot
plt.show()
```

# Customizing Seaborn Plots


```python
#Figure style includes background and axes
#Preset options : "white","dark","whitegrid","darkgrid","tick"
#sns.set_style()
#sns.set_palette() 
#1) Diverging palette - [Default palettes : "RdBu" , "PrGn" , "RdBu_r" , "PrGn_r" (reverse of the latter)]
#2) Sequential palette - "Greys" , "Blues" , "PuRd" , "GnBu"
#3) Custom_palette = [col1, col2]
#change the scale
#sns.set_context() : paper,notebook,talk,poster
```

```python
# Set the style to "Whitegrid"
# Set the color palette to "Purples"
sns.set_style("whitegrid")
sns.set_palette("Purples")

# Create a count plot of survey responses
category_order = ["Never", "Rarely", "Sometimes", 
                  "Often", "Always"]

sns.catplot(x="Parents' advice", 
            data=survey_data, 
            kind="count", 
            order=category_order)

# Show plot
plt.show()
```

```python
# Set the context to "paper"
sns.set_context("paper")

# Create bar plot
sns.catplot(x="Siblings", y="Loneliness",
            data=survey_data, kind="bar")

# Show plot
plt.show()
```

```python
# Set the style to "darkgrid"
sns.set_style("darkgrid")

# Set a custom color palette
sns.set_palette(["#39A7D0","#36ADA4"])

# Create the box plot of age distribution by gender
sns.catplot(x="Gender", y="Age", 
            data=survey_data, kind="box")

# Show plot
plt.show()
```

### Adding titles and labels: Part 1

```python
# Seaborn plots creates two different type of objects
#a) FacetGrid(can create subplots such as relplot(),catplot()) 
#b) AxesSubplot (only create a single plot such as scatterplot(), countplot())
```

### Adding title to facetgrid object

```python
# Create scatter plot
g = sns.relplot(x="weight", 
                y="horsepower", 
                data=mpg,
                kind="scatter")

# Identify plot type
type_of_g = type(g)

# Print type
print(type_of_g)
```

```python
# Create scatter plot
g = sns.relplot(x="weight", 
                y="horsepower", 
                data=mpg,
                kind="scatter")

# Add a title "Car Weight vs. Horsepower"
g.fig.suptitle("Car Weight vs. Horsepower", y=1.05)

# Show plot
plt.show()
```

### Adding title to a AxesSubplot object

```python
# Create line plot
g = sns.lineplot(x="model_year", y="mpg", 
                 data=mpg,
                 hue="origin")

# Add a title "Average MPG Over Time"
g.set_title("Average MPG Over Time")

# Show plot
plt.show()
```

```python
# Create point plot
sns.catplot(x="origin", 
            y="acceleration", 
            data=mpg, 
            kind="point", 
            join=False, 
            capsize=0.1)

# Rotate x-tick labels
plt.xticks(rotation=90)

# Show plot
plt.show()
```
